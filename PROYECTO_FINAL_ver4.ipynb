```python
plantilla1='''

               |
               |
               |
               |
               |'''


plantilla2='''
          
                            
                ___________
               | /        
               |/        
               |              ¡vamos tu puedes! inentalo de nuevo
               |          
               |'''


plantilla3='''
          
                ___________
               | /        | 
               |/              ¡rayos! volviste a fallar 
               |          
               |          
               |                                                                                                                                                      
    '''


plantilla4='''


                ___________
               | /        |        ¡oh, ho!
               |/        ( )
               |          
               |          
               |                                                                                                                                                
 '''              


plantilla5='''
                ___________
               | /        | 
               |/        ( )      ¡esto se pone interesante!
               |          |
               |          
               |             
   '''  


plantilla6='''
                ___________
               | /        | 
               |/        ( )           ¡qué meyo!
               |          |
               |         / 
               |                    
  '''


plantilla7='''
                 ___________
               | /        | 
               |/        ( )          ¡auch!
               |          |
               |         / \\
               |   
               |      
  '''



palabra_secreta= 'piel'
letras_ingresadas=['_','_','_','_']    #para meter las letras ingresadas en una lista
lista_palabra_secreta=[] #crear una lista vacia para meter la palabra secreta
lista_palabra_secreta.extend(palabra_secreta)   #meter la palabra secreta en la lista
ahorcado={1:plantilla1, 2:plantilla2, 3:plantilla3, 4:plantilla4, 5:plantilla5, 6:plantilla6, 7:plantilla7}
entrada=1
m=int(0)   #registro de intentos

longitud=len(lista_palabra_secreta)
vidas=int(7)
print(f'''
{30*'= '}
Bien venido al juego del ahorcado:
{30*'= '}''')
print(f'''
Adivina la palabra secreta:''')
print(f'''{"__ "  * len(palabra_secreta)} ({len(palabra_secreta)} letras)''')
while longitud>0 and vidas >0:                
  
  print(f'''
Te quedan {vidas} vidas.
        ''')
  print(f'''------------------------------------------------
  ''')
  
  letra=input(f'''Ingresa una letra (en minúscula): ''')
  letra=letra.lower()
  print(f' ')
  if letra in letras_ingresadas:
    print('Esa letra ya la adivinaste (este intento no contará')
  else:
    m=m+1
    if(letra in lista_palabra_secreta):
      
      print(f'''La letra "{letra}"" sí está en la palabra secreta y va colocada así:
      ''' )
      
      while letra in lista_palabra_secreta:
        
        indice=int(lista_palabra_secreta.index(letra))
        lista_palabra_secreta[indice]='_'
        letras_ingresadas[indice]=letra
 
      print(letras_ingresadas)
      longitud=longitud-palabra_secreta.count(letra)
 
    else:
      print(f'''La letra "{letra}"" no está en la palabra secreta''')
      vidas=vidas-1
#*********************************************
      if entrada<=7:
        print(' ')
        print(ahorcado[entrada])
        entrada=entrada+1
#*********************************************
    
    
if vidas==0:
  print(f'''\n¡Lastima! se acabaron tus oportunidades la palabra era "{palabra_secreta}"''')
else:
  print(f'''\n¡Felicidades! lo lograste en {m} intentos, la palabra secreta es "{palabra_secreta}"''')




```
